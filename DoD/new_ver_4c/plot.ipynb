```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 200
```

```python
labels = [500, 1000, 2000]
time_500 = [2.932798147201538, 5.243244171142578, 1057.8104031085968, 1373.2922151088715]
time_1000 = [6.028142213821411, 9.870358228683472, 3641.072788000107, 6624.540374040604]
time_2000 = [11.925531148910522, 16.36414098739624, 21033.039329767227, None]

compatible = [2.932798147201538, 6.028142213821411, 11.925531148910522]
contained = [5.243244171142578, 9.870358228683472, 16.36414098739624]
contradictory = [1057.8104031085968, 3641.072788000107, 21033.039329767227]
complementary = [1373.2922151088715, 6624.540374040604, np.nan]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - 3 * width/2, compatible, width, label='compatible')
rects2 = ax.bar(x - width/2, contained, width, label='contained')
rects3 = ax.bar(x + width/2, contradictory, width, label='contradictory')
rects4 = ax.bar(x + 3 * width/2, complementary, width, label='complementary')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('run time(s)')
ax.set_xlabel('num views')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=2)
ax.bar_label(rects2, padding=2)
ax.bar_label(rects3, padding=2)
ax.bar_label(rects4, padding=2)

ax.set_yscale('log')

fig.tight_layout()

plt.show()
```
